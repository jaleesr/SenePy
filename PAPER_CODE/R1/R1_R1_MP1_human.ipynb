```python
@nb.njit
def increasing_logistic_function(x, k, x0):
    return 1 / (1 + np.exp(-k * (x - x0)))

@nb.njit
def starting_function(x):
    x = x*100
    if x < 5:
        return 1

    return -x/2 + 3.5

@nb.njit
def max_function(x):
    x = x*100
    if x <= 0:
        return 0
    elif 0 < x <= 3:
        #return (1/3)*x
        return increasing_logistic_function(x, 2, 0)
    elif 3 < x <= 20:
        return 1
    elif 20 < x <= 100:
        return -(1/4)*x + 6 #y = 1 at 20
    else:
        return 0
    
@nb.njit
def gain_function(x):
    x = x*100
    
    if x < 5:
        return x/5
    
    if x > 15:
        return -x/5 + 4

    return 1
    
@nb.njit
def max_v_end_function(x):
    x = x*100
    if x < 50:
        return 1
    return -(x - 50)

@nb.njit
def max_age_function(x):
    if x >= 38:
        return 1
    return x - 38




increasing_logistic_function(1,1,1)
starting_function(1)
max_function(1)
gain_function(1)
max_v_end_function(1)
max_age_function(1)
```

```python
def soft_stats(df):
    df = df.copy()
    
    df['max_v_end'] = df.Max - df.ending
    # try:
    #     df['max_age'] = df[['8','18','28','38','48','58','68','78','88']].idxmax(axis = 1)
    # except:
    #     df['max_age'] = df[[8,18,28,38,48,58,68,78,88]].idxmax(axis = 1)
    if '8' in df.columns:
        df['max_age'] = df[['8','18','28','38','48','58','68','78','88']].idxmax(axis = 1)
    else:
        df['max_age'] = df[[8,18,28,38,48,58,68,78,88]].idxmax(axis = 1)
    
    df['max_age'] = df['max_age'].astype(int)
    
    
    df['soft_start'] = np.vectorize(starting_function)(df['starting'].values)
    df['soft_max'] = np.vectorize(max_function)(df['Max'].values)
    df['soft_gain'] = np.vectorize(gain_function)(df['gain'].values)
    df['soft_max_age'] = np.vectorize(max_age_function)(df['max_age'].values)
    df['soft_mve'] = np.vectorize(max_v_end_function)(df['max_v_end'].values)
    df['soft_slope'] = df.slope * 500 #try normalization instead

    df['metric'] = df[['soft_start', 'soft_max', 'soft_gain', 'soft_max_age', 'soft_mve', 'soft_slope']].sum(axis = 1)
    return df


def calculate_p_values(null_array, test_values):
    sorted_null = np.sort(null_array)
    indices = np.searchsorted(sorted_null, test_values, side='right')
    p_values = (N - indices) / N
    return p_values
```

```python
##OG filtering
def extra_filt(x):
    x = x[x >= 0]
    if x.max() > 2*x.iloc[-1] or x.idxmax() <= 38: # 
        return False
    else:
        return True

def origi_filt(df):
    cf = df.copy()

    cf = cf[(cf.slope > 0) &\
      (cf.Max < 0.2) &\
      (cf.starting < .05) &\
      (cf.gain > 0.015)]

    cf = cf[cf[[8,18,28,38,48,58,68,78,88]].apply(extra_filt, axis = 1)]

    return df.gene.isin(cf.gene)
    
```

```python
res = []
for tissue, cell in tqdm(input_list):
    s = 'permutation_output_human/' + tissue + '.' + cell + '.feather'
    nulls = pd.read_feather(s)
    nulls['gain'] = nulls.ending - nulls.starting

    N = len(nulls)
    nulls = soft_stats(nulls)

    #load test data
    a = tc_df[(tc_df.tissue == tissue) &\
                (tc_df.cell_type_2 == cell)].copy()
    a = soft_stats(a)
    
    a['p'] = calculate_p_values(nulls['metric'].to_numpy(), a['metric'].to_numpy())
    a['OrigFilt'] = origi_filt(a)
    
    a['tissue_cell'] = tissue + '.' + cell
    
    res.append(a)
```

```python
sel = {}
dd = 0
for c in df.tissue_cell.unique():
    _ = df[df.tissue_cell == c]
    _ = _.sort_values('metric', ascending = False)
    _['BH'] = _['p'] * len(_) / (_.reset_index().index + 1)
    _['BH'] = _['BH'].clip(upper=1)
    _ = _.reset_index(drop = True)
    selected = _[(_.BH < 0.05) | ((_.OrigFilt) & (_.p < .01))]
    a = set(_[_.OrigFilt].gene)
    b = set(selected.gene)
    i = a.intersection(b)
    d = len(b) - len(a)

    dd += d
    
    p = stats.hypergeom.sf(len(i), len(_), len(a), len(b))

    sel[c] = list(b)
    
    print(f'{len(a)}\t{len(b)}\t{len(i)}\t{d}\t{p}')
```
