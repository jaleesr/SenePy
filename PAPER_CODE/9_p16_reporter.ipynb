```python
#returns df of correlations between hubs and the td counts
def correlate_it(adata, cell):
    out = []
    for column in adata.obsm['hub_scores'].columns:
        sub = adata[(adata.obs['cell type'] == cell) & (adata.obs.tom == 'tom+')]
        a = sub.obs['tom UMI']
        b = sub.obsm['hub_scores'][column]
        res = stats.pearsonr(a,b)
        out.append([cell, column, res[0], res[1]])
        
    df = pd.DataFrame(out, columns = ['cell', 'hub', 'r', 'p'])
    df = df.sort_values('p').reset_index(drop = True)
    df['BH'] = df.p * len(df) / (df.index + 1)
    return df


#returns df of correlations between hubs and the td counts
def correlate_it_plot(adata, cell, hub):

    sub = adata[(adata.obs['cell type'] == cell) & (adata.obs.tom == 'tom+')]
    a = sub.obs['tom UMI']
    b = sub.obsm['hub_scores'][hub]
        
    df = pd.DataFrame([a,b]).T
    
    ax = sns.regplot(data = df, x = 'tom UMI', y = hub)
        
        
    #return df

#returns df of correlations between hubs and if cell is p16+ or not
#doesnt seem as powerful as above
def correlate_binary(adata, cell):
    out = []
    for column in adata.obsm['hub_scores'].columns:
        sub = adata[(adata.obs['cell type'] == cell)]
        a = sub.obs.tom == 'tom+' #binarize
        b = sub.obsm['hub_scores'][column]
        res = stats.pearsonr(a,b)
        out.append([cell, column, res[0], res[1]])

    df = pd.DataFrame(out, columns = ['cell', 'hub', 'r', 'p'])
    df = df.sort_values('p').reset_index(drop = True)
    df['BH'] = df.p * len(df) / (df.index + 1)
    return df


def test_distribution(adata, cell):
    out = []
    for column in adata.obsm['hub_scores'].columns:
        sub = adata[(adata.obs['cell type'] == cell)]
        sub.obs['the_score'] = sub.obsm['hub_scores'][column]
        
        
        a = sub[sub.obs.tom == 'tom+'].obs.the_score.values
        
        b = sub[sub.obs.tom == 'tom-'].obs.the_score.values
        
        res = stats.mannwhitneyu(a,b)
        out.append([cell, column, res.statistic, res.pvalue, a.mean(), b.mean()])

    df = pd.DataFrame(out, columns = ['cell', 'hub', 'stat', 'p', 'tom+', 'tom-'])
    df = df.sort_values('p').reset_index(drop = True)
    df['BH'] = df.p * len(df) / (df.index + 1)
    df['dif'] = df['tom+'] - df['tom-']
    return df
```
